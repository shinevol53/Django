#### 1. MVC & MTV

- M: model(안전하게 데이터를 저장)
- V: view(데이터를 적절하게 유저에게 보여줌) - 디자이너
- C: control / T: template(Django)
    - (사용자의 입력과 이벤트에 반응하여 Model과 View를 업데이트)

#### 2. Django개념(특정 역할 분리 위해)

web browser->URL Dispatcher -> View -> Model -> DataBase -> Model -> View
-> Template(javascript, html, css) -> web browser

browser -> web server(Nginx/Apache) -> WSGI(wsgi.py) -> URL Resolution(urls.py)
-> VIEW(views.py) -> MODEL(models.py) -> VIEW(views.py):가공
-> TEMPLATE(example.html):script파일, 다양한 form작성 ->WSGI
-> web server

- FORM(forms.py): 모델과 템플레이트 있는 사용자들이 쓰는 UI가 쉽게 관리가 된다. 굉장히 편리하게 장고에서 구성이 된다.

- database query 만들어서 변수에 넣고 하는 과정을 Django는
- 모델에 변수만 지정하면 뒤에서 데이터베이스와 연결하는
- MANAGERS가 알아서 작성
- (so, sql 작성할 필요 없음)

#### 4. Settings.py(프로젝트 환경 설정 파일)
- DEBUG
: 디버그 모드 설정
: true로 놓으면 다양한 변수 활용
: 서비스 런칭하면 false로 돌려야 함

- INSTALLED_APPS
: pip로 설치한 앱 또는 본인이 만든 app 추가

- MIDDLEWARE_CLASSES
: request와 response 사이의 주요 기능 레이어
: 인증/보완 관련 기능이 있음

- TEMPLATES
: django template관련 설정, 실제 뷰(html, 변수)

- DATABASES
: 데이터베이스 엔진의 연결 설정

- STATIC_URL
: 정적 파일의 URL(css, javascript, image, ect.)

#### 5. manage.py
- 프로젝트 관리 명령어 모음
- 주요 명령어
 - startapp = 앱 설정
 - runserver = 서버 실행
 - createsuperuser = 관리자 생성
 - makemigrations app = app의 모델 변경 사항 체크
 - migrate = 변경 사항을 DB에 반영
 - shell = 쉘을 통해 데이터를 확인
 - collectstatic = static 파일을 한 곳에 모음

ex) ./manage.py runserver 0.0.0.0:8080
